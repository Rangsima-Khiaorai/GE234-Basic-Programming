
## **2. การจัดการข้อมูลตารางด้วย Pandas**
### **2.1 สร้าง DataFrame ตัวอย่าง**
ลองสร้าง DataFrame ข้อมูลประชากรของประเทศในภูมิภาค ASEAN
    

```python

import pandas as pd

# สร้าง DataFrame ตัวอย่าง
data = {
    'ประเทศ': ['ไทย', 'ลาว', 'กัมพูชา', 'เวียดนาม', 'มาเลเซีย', 'อินโดนีเซีย', 'สิงคโปร์', 'ฟิลิปปินส์', 'บรูไน', 'เมียนมาร์'],
    'ประชากร (ล้าน)': [69.8, 7.1, 16.5, 97.3, 33.5, 277.5, 5.7, 117.3, 0.45, 54.2],
    'พื้นที่ (ตร.กม.)': [513120, 236800, 181035, 331210, 330803, 1904569, 728, 300000, 5765, 676578]
}
df = pd.DataFrame(data)

# แสดงข้อมูล
print(df)

# คำนวณความหนาแน่นประชากร
df['ความหนาแน่นประชากร (คน/ตร.กม.)'] = df['ประชากร (ล้าน)'] * 1_000_000 / df['พื้นที่ (ตร.กม.)']
print(df)
```

## **3. การจัดการข้อมูลภูมิศาสตร์ด้วย GeoPandas** 5 คะแนน
### **3.1 การอ่าน Shapefile และแสดงผล**
ใช้ข้อมูล Shapefile ที่มีขอบเขตและมีประชากร อาจจะหาจากแหล่งข้อมูลที่ เป็น Opendata เช่น
https://data.humdata.org/dataset/worldpop-population-density-for-thailand

    

```python

import geopandas as gpd

# อ่านไฟล์ Shapefile
shapefile_path = '/content/drive/MyDrive/GE234/populationthailand.shp'
geo_data = gpd.read_file(shapefile_path)

# แสดงข้อมูลเบื้องต้น
print(geo_data.head())
print(geo_data.crs)

# แสดงผลข้อมูลบนแผนที่
geo_data.plot(column='pop_total', cmap='OrRd', legend=True)
```


## **4. การแสดงผลข้อมูล**
### **4.1 สร้างกราฟด้วย Matplotlib**  5 คะแนน
สร้างกราฟแสดงประชากรของแต่ละประเทศในภูมิภาค ASEAN โดยให้นักศึกษาทดลองหาข้อมูลเอง
    

```python
data = {
    'Country': ['Thailand', 'Laos', 'Cambodia', 'Vietnam', 'Malaysia', 'Indonesia', 'Singapore', 'Philippines', 'Brunei Darussalam', 'Myanmar'],
    'Population (M)': [69.8, 7.1, 16.5, 97.3, 33.5, 277.5, 5.7, 117.3, 0.45, 54.2],
    'Area (qr km)': [513120, 236800, 181035, 331210, 330803, 1904569, 728, 300000, 5765, 676578]
}
df = pd.DataFrame(data)

import matplotlib.pyplot as plt

# ปรับขนาดกราฟให้เหมาะสม
plt.figure(figsize=(12, 6))

# สร้างกราฟแสดงประชากร
plt.bar(df['Country'], df['Population (M)'], color='skyblue')
plt.title('Population of each country')
plt.xlabel('country')
plt.ylabel('Population (M)')
plt.xticks(rotation=45, ha='right') # หมุนฉลากแกน x เพื่อให้อ่านง่ายขึ้น
plt.tight_layout() # ปรับเลย์เอาต์เพื่อป้องกันฉลากทับซ้อนกัน
plt.show()
```


### **4.2 การรวมข้อมูลเชิงพื้นที่และตาราง**
รวมข้อมูล Shapefile กับข้อมูลประชากรเพื่อแสดงข้อมูลประชากรในแต่ละจังหวัด โดยให้นักศึกษาทดลองหาข้อมูลเอง
    

```python

# รวมข้อมูล
merged_data = geo_data.merge(df, left_on='province', right_on='ประเทศ')

# แสดงผลข้อมูลในแผนที่
merged_data.plot(column='ประชากร (ล้าน)', cmap='OrRd', legend=True, figsize=(12, 8))
plt.title('ประชากรในแต่ละจังหวัด')
plt.show()
```

ข้อ 3

```python
# สร้างแผนที่แสดง 'area' ของแต่ละอำเภอในจังหวัดกาญจนบุรี
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
geo_data.plot(column='area', cmap='GnBu', legend=True,
              legend_kwds={'label': "พื้นที่ (ตร.ม.)",
                           'orientation': "vertical"})
plt.title('แผนที่แสดงพื้นที่ (ตร.ม.) ของแต่ละอำเภอ')
plt.xlabel('ลองจิจูด')
plt.ylabel('ละติจูด')
plt.show()
```

```python
# คำนวณความหนาแน่นประชากรในจังหวัดกาญจนบุรี
# ตรวจสอบว่าคอลัมน์ 'area' มีค่าเป็น 0 หรือไม่ เพื่อหลีกเลี่ยงการหารด้วยศูนย์
geo_data['density'] = geo_data.apply(lambda row: row['NOPEOPLE'] / row['area'] if row['area'] > 0 else 0, axis=1)

# สร้างแผนที่แสดง 'density' ของแต่ละอำเภอ
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
geo_data.plot(column='density', cmap='YlOrRd', legend=True,
              legend_kwds={'label': "ความหนาแน่นประชากร (คน/ตร.ม.)",
                           'orientation': "vertical"})
plt.title('แผนที่แสดงความหนาแน่นประชากร (คน/ตร.ม.) ของแต่ละอำเภอ')
plt.xlabel('ลองจิจูด')
plt.ylabel('ละติจูด')
plt.show()
```
