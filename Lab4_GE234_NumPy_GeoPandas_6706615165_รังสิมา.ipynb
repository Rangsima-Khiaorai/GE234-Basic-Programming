## 🔹 ตัวอย่างที่ 1: ใช้ NumPy คำนวณข้อมูล NDVI จากภาพ Raster


```python

import numpy as np

# สร้างข้อมูลตัวอย่างสำหรับภาพ NDVI (Normalized Difference Vegetation Index)
nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])

# คำนวณค่า NDVI
ndvi = (nir - red) / (nir + red)

print("ค่า NDVI:")
print(ndvi)
```


## 🔹 ตัวอย่างที่ 2: ใช้ NumPy คำนวณค่าเฉลี่ยและค่ามากสุดของ NDVI


```python

print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.2f}")
```


## 🔹 ตัวอย่างที่ 3: ใช้ GeoPandas โหลดและวิเคราะห์ข้อมูล Shapefile


## 🔹 ตัวอย่างที่ 4: คำนวณพื้นที่ของแต่ละจังหวัด


```python

# ตรวจสอบค่า CRS (Coordinate Reference System)
print(gdf.crs)

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางเมตร)
gdf["area_sqkm"] = gdf.geometry.area / 1e6

# แสดงพื้นที่จังหวัด 5 อันดับแรกที่ใหญ่ที่สุด
print(gdf.nlargest(5, "area_sqkm")[["PROVINCE_T", "area_sqkm"]])
```

### 1. **แบบฝึกหัด 1:** ใช้ NumPy คำนวณค่า Mean, Max, Min ของค่า NDVI ในอาร์เรย์ที่สร้างขึ้นเอง

```python
import numpy as np

# สร้างข้อมูลตัวอย่างสำหรับภาพ NDVI (Normalized Difference Vegetation Index)
nir = np.array([[0.9, 0.3, 0.7], [0.4, 0.7, 0.8], [0.5, 0.6, 0.9]])
red = np.array([[0.6, 0.1, 0.5], [0.2, 0.4, 0.6], [0.2, 0.4, 0.6]])

# คำนวณค่า NDVI
ndvi = (nir - red) / (nir + red)

print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.2f}")
```



---



### 2. **แบบฝึกหัด 2:** ใช้ GeoPandas โหลด Shapefile ของจังหวัด และคำนวณพื้นที่ของแต่ละจังหวัด

```python
import geopandas as gpd

# โหลดข้อมูลชั้นข้อมูลจังหวัดของประเทศไทย พร้อมระบุ encoding เพื่อให้แสดงภาษาไทยได้ถูกต้อง
gdf = gpd.read_file("/content/drive/MyDrive/Thailand/Province.shp", encoding='cp874')

# ตรวจสอบค่า CRS (Coordinate Reference System)
print(gdf.crs)

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางกิโลเมตร)
gdf["area_sqkm"] = gdf.geometry.area / 1e6

# แสดงพื้นที่ของทุกจังหวัด
display(gdf[["PROVINCE_T", "area_sqkm"]])
```



---



### 3. **แบบฝึกหัด 3:** ใช้ GeoPandas ทำ Spatial Join ระหว่างข้อมูลจังหวัดและอำเภอ

```python
import geopandas as gpd
import pandas as pd

# โหลดข้อมูลชั้นข้อมูลจังหวัดของประเทศไทย พร้อมระบุ encoding เพื่อให้แสดงภาษาไทยได้ถูกต้อง
gdf = gpd.read_file("/content/drive/MyDrive/Thailand/Province.shp", encoding='cp874')

# โหลดข้อมูลชั้นข้อมูลอำเภอ
gdf_districts = gpd.read_file("/content/drive/MyDrive/Thailand/Amphoe.shp", encoding='cp874')

# ทำ Spatial Join ระหว่างอำเภอกับจังหวัด
gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

# ปรับการตั้งค่าการแสดงผลของ pandas เพื่อลดความกว้างของคอลัมน์ที่มีข้อความยาวๆ
pd.set_option('display.max_colwidth', 50) # สามารถปรับค่าตัวเลขนี้ได้ตามความเหมาะสม

# แสดงข้อมูลที่เชื่อมโยงกันทั้งหมด โดยไม่แสดงคอลัมน์ 'geometry' เนื่องจากข้อจำกัดด้านพื้นที่ในตาราง เพื่อทำให้ตารางอ่านง่ายขึ้น
display(gdf_joined.drop(columns=['geometry']))

# เพื่อให้การตั้งค่ากลับเป็นค่าเริ่มต้นหลังจากแสดงผล
pd.reset_option('display.max_colwidth')
```

### 4. **แบบฝึกหัด 4:** ใช้ NumPy และ GeoPandas ร่วมกันเพื่อหาข้อมูลจังหวัดที่มี NDVI เฉลี่ยสูงสุด

```python
import numpy as np
import pandas as pd

# ตรวจสอบให้แน่ใจว่า gdf (ข้อมูลจังหวัด) ถูกโหลดไว้แล้ว
# ถ้ายังไม่ได้โหลด ให้รันเซลล์ที่โหลดข้อมูลจังหวัดก่อนหน้านี้ (เช่น เซลล์ในแบบฝึกหัด 2 หรือ 3)
if 'gdf' not in locals():
    import geopandas as gpd
    print("Loading gdf again as it was not found in current session...")
    gdf = gpd.read_file("/content/drive/MyDrive/Thailand/Province.shp", encoding='cp874')

# จำลองค่า NIR (Near-Infrared) และ Red (Visible Red) สำหรับแต่ละจังหวัด
# ค่า NIR มักจะสูงกว่า Red สำหรับพื้นที่ที่มีพืชพรรณหนาแน่น
np.random.seed(42) # เพื่อให้ผลลัพธ์จำลองคงที่เมื่อรันซ้ำ
nir_values = np.random.uniform(low=0.4, high=0.9, size=len(gdf))
red_values = np.random.uniform(low=0.1, high=0.4, size=len(gdf))

# คำนวณ NDVI สำหรับแต่ละจังหวัด
# NDVI = (NIR - Red) / (NIR + Red)
# จัดการกรณีที่ NIR + Red เป็นศูนย์เพื่อป้องกันการหารด้วยศูนย์
sum_bands = nir_values + red_values
calculated_ndvi = np.where(sum_bands == 0, 0, (nir_values - red_values) / sum_bands)

# เพิ่มค่า NIR, Red และ NDVI ที่คำนวณได้เข้าไปใน GeoDataFrame
gdf['simulated_nir'] = nir_values
gdf['simulated_red'] = red_values
gdf['calculated_ndvi'] = calculated_ndvi

# หาจังหวัดที่มีค่า NDVI สูงสุด
province_max_ndvi = gdf.loc[gdf['calculated_ndvi'].idxmax()]

print("\n--- ผลการจำลองค่า NDVI และจังหวัดที่มี NDVI สูงสุด ---")
print(f"จังหวัดที่มีค่า NDVI เฉลี่ยสูงสุด (จำลอง): {province_max_ndvi['PROVINCE_T']}")
print(f"ค่า NDVI จำลองสูงสุด: {province_max_ndvi['calculated_ndvi']:.2f}")

# แสดงข้อมูลจังหวัดพร้อมค่า NIR, Red และ NDVI ที่คำนวณได้ทั้งหมด
print("\nข้อมูลจังหวัดพร้อมค่า NIR, Red และ NDVI ที่คำนวณได้ทั้งหมด:")
display(gdf[['PROVINCE_T', 'simulated_nir', 'simulated_red', 'calculated_ndvi']])
```
